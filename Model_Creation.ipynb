```python
import pandas as pd
import numpy as np
pd.options.display.max_columns=None
```

```python
df=pd.read_csv('preprocessed_weather.csv')
```

```python
df.head(10)
```

```python
df[df['RainTomorrow']==1]
```

```python
df.isnull().sum()
```

```python
x = df.drop(["RainTomorrow", "Date"], axis=1)
y = df["RainTomorrow"]
```

```python
len(x.columns)
```

```python
x.head()
```

```python
x.dtypes
```

```python
convert_dict={'Location':'float64','RainToday':'float64','Month':'float64','Day':'float64','Year':'float64'}
```

```python
x=x.astype(convert_dict)
```

```python
x.dtypes
```

```python
y=y.astype('float64')
```

```python
y.dtype
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(xtrain)
```

```python
xtrain=scaler.transform(xtrain)
```

```python
xtrain.shape
```

```python
xtest=scaler.transform(xtest)
```

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression(random_state=0)
model1.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(ytest,model1.predict(xtest)))
print(confusion_matrix(ytest,model1.predict(xtest)))
print(classification_report(ytest,model1.predict(xtest)))
```

```python
from catboost import CatBoostClassifier
model2 = CatBoostClassifier(iterations=2000, eval_metric = "AUC")
model2.fit(xtrain,ytrain)
```

```python
print(accuracy_score(ytest,model2.predict(xtest)))
print(confusion_matrix(ytest,model2.predict(xtest)))
print(classification_report(ytest,model2.predict(xtest)))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model2, xtrain, ytrain, cv = 5, scoring='accuracy')
print('Cross-validation scores:{}'.format(scores))
print('Average cross-validation score: {}'.format(scores.mean()))
```

```python
import pickle
with open('model.pickle','wb') as f:
  pickle.dump(model2,f)
```

```python
with open('scaler.pickle','wb') as f:
  pickle.dump(scaler,f)
```
